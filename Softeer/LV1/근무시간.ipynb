```python
def time_cal(start_h, start_m, end_h, end_m) :
    work_time =0

    if start_m > end_m :
        end_h -= 1
        end_m += 60
    
    work_h = end_h - start_h
    work_m = end_m - start_m

    work_time = work_h * 60 + work_m

    return work_time

def split(str_time) :
    start, end = str_time.split()

    start_h, start_m = map(int, start.split(":"))
    end_h, end_m = map(int, end.split(":"))

    work_time = time_cal(start_h, start_m, end_h, end_m)
    return work_time

tot = 0

for _ in range(5) :
    tot += split(input())

print(tot)
```
