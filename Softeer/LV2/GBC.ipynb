```python
n, m = map(int, input().split())

speed = [0] * 101
idx = 1

for i in range(n) :
    rule_len, rule_val = map(int, input().split())
    for j in range(idx, idx+rule_len) :
        speed[j] = rule_val
    
    idx = idx + rule_len

idx = 1
result = 0

for i in range(m) :
    test_len, test_val = map(int, input().split())
    for j in range(idx, idx+test_len) :
        result = max(result, test_val - speed[j])
    
    idx = idx + test_len

print(result)
```
