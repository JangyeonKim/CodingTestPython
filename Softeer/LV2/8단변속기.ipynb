```python
gear = list(map(int, input().split()))

a = [x for x in range(1,9)]

# d = a.sort(reverse=True) 
d = sorted(a, reverse=True)

if gear == a :
    print("ascending")
elif gear == d :
    print("descending")
else :
    print("mixed")
```

d = a.sort() 는 잘못된 문법.
a.sort() 는 a 자체를 정렬시키는 것이고, 정렬시킨 후에 None을 반환한다.
