```python
def cal_metal_weight(metal_weight, w) :
    each_metal_weight = [] # price     

    for weight in metal_weight :
        if w > weight : #            .
            each_metal_weight.append(weight)
            w -= weight
        else : #          .
            each_metal_weight.append(w)
            break
    
    return each_metal_weight


w, n = map(int, input().split())

metal_info = []

for i in range(n) :
    metal_info.append(list(map(int, input().split())))
    
metal_info.sort(key = lambda x : x[1], reverse = True)
# print(metal_info) [[70,2],[90,1]]

metal_price = [x[1] for x in metal_info]
# print(metal_price) [2, 1]
metal_weight = [x[0] for x in metal_info]
# print(metal_weight) [70, 90]

each_metal_weight = cal_metal_weight(metal_weight, w)
# print(each_metal_weight) [70, 30]

price = 0

for i in range(len(each_metal_weight)) :
    price += metal_price[i] * each_metal_weight[i]

print(price)
```
