- 탑다운 방식 : 구현할 때 재귀함수를 이용, 큰 문제를 해결하기 위해 작은 문제를 재귀적으로 호출
- 보텀업 방식 : 반복문을 이용, 작은 문제부터 차근차근 답을 도출하고 이를 활용하여 큰 문제를 해결

#### <다이나믹 프로그래밍을 사용하는 대표적인 예시>

```python
# 피보나치 수열 : 재귀적으로 구현

def fibo(n):
    if n == 1 or n == 2: # n이 2보다 작거나 같으면 1을 반환
        return 1
    return fibo(n-1) + fibo(n-2) # n이 2보다 크면 앞쪽 두 항의 합을 반환

print(fibo(10)) # 구하려는 피보나치 수열의 항에 따라 시간이 기하급수적으로 증가한다.
```

```python
# 피보나치 수열 : 탑다운 다이나믹 프로그래밍으로 구현

d = [0] * 100 # 한 번 계산된 결과를 메모이제이션하기 위한 리스트 초기화

def fibp_td(x) :
    if x==1 or x == 2 : # 종료 조건
        return 1
    if d[x] != 0 : # 이미 계산한 적 있는 문제라면 그대로 반환
        return d[x]
    d[x] = fibp_td(x-1) + fibp_td(x-2) # 아직 계산하지 않은 문제라면 점화식에 따라서 피보나치 결과 반환
    return d[x]

print(fibp_td(99))
```

```python
# 피보나치 수열 : 보텀업 다이나믹 프로그래밍으로 구현

d = [0] * 100

d[1] = 1
d[2] = 1
n = 99

# 피보나치 함수 반복문으로 구현
for i in range(3, n+1) :
    d[i] = d[i-1] + d[i-2] 
    
print(d[n])
```
