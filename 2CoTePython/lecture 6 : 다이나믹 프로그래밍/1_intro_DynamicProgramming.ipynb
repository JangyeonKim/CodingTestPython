```python
#   :  

def fibo(n):
    if n == 1 or n == 2: # n 2   1 
        return 1
    return fibo(n-1) + fibo(n-2) # n 2      

print(fibo(10)) #        .
```

```python
#   :    

d = [0] * 100 #        

def fibp_td(x) :
    if x==1 or x == 2 : #  
        return 1
    if d[x] != 0 : #       
        return d[x]
    d[x] = fibp_td(x-1) + fibp_td(x-2) #         
    return d[x]

print(fibp_td(99))
```

```python
#   :    

d = [0] * 100

d[1] = 1
d[2] = 1
n = 99

#    
for i in range(3, n+1) :
    d[i] = d[i-1] + d[i-2] 
    
print(d[n])
```
