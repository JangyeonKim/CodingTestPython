##### <문제 1 : 시각>



```python
import time

N = int(input())

s_t = time.time()

cnt = 0

for h in range(N+1) : 
    for m in range(60) :
        for s in range(60) :
            t = str(h) + str(m) + str(s)
            if '3' in t :
                cnt += 1
                
print(cnt)

e_t = time.time()
print("time :", e_t - s_t)
```

##### <문제 2 : 왕실의 나이트>





```python
import time

loc = input()

s_t = time.time()

col_dict = {'a' : 0, 'b' : 1, 'c' : 2, 'd' : 3, 'e' : 4, 'f' : 5, 'g' : 6, 'h' : 7}

row, col = int(loc[1])-1, col_dict[loc[0]]

drow = [1, 1, -1, -1, 2, 2, -2, -2]
dcol = [2, -2, 2, -2, 1, -1, 1, -1]

cnt = 8

for i in range(8) :
    rr = row + drow[i]
    cc = col + dcol[i]
    
    if rr > 7 or cc > 7 or rr < 0 or cc < 0 :
        cnt -= 1

print(cnt)

e_t = time.time()
print("time :", e_t - s_t)
```
