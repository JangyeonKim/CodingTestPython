```python
def process(n, k, cnt=0) :
    while n != 1 : # n 1  
        if n % k != 0  : # n k   
            n -= 1
        else : # n k  
            n /= k
            
        cnt+=1 #    
    
    return cnt


N, K = map(int, input().split()) 
iter_num = process(N, K)

print(iter_num)
```

```python
s = input()

result = int(s[0])

for i in range(1, len(s)) :
    num = int(s[i])
    
    if num <=1 or result<=1 : #     0 or 1  
        result += num
    else :
        result *= num
        
print(result)
```
