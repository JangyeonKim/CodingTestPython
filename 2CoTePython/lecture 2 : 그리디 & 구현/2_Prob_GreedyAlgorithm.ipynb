##### <문제1 : 1이 될 때까지>




```python
def process(n, k, cnt=0) :
    while n != 1 : # n이 1이 아닐동안 반북
        if n % k != 0  : # n이 k로 나누어 떨어지지 않으면
            n -= 1
        else : # n이 k로 나누어 떨어지면
            n /= k
            
        cnt+=1 # 반복될 때마다 횟수 증가
    
    return cnt


N, K = map(int, input().split()) 
iter_num = process(N, K)

print(iter_num)
```

##### <문제2 : 곱하기 혹은 더하기>




```python
s = input()

result = int(s[0])

for i in range(1, len(s)) :
    num = int(s[i])
    
    if num <=1 or result<=1 : # 두 수 중 하나라도 0 or 1인 경우 더하기
        result += num
    else :
        result *= num
        
print(result)
```

##### <문제3 : 모험가 길드>




```python
N = int(input()) # 5
fear_list = list(map(int, input().split())) # [2, 3, 1, 2, 2]

fear_list.sort() # [1, 2, 2, 2, 3]

result = 0 # 그룹 수
count = 0 # 현재 그룹에 포함된 인원 수

for i in fear_list :
    count += 1 # 그룹의 인원 수 증가 
    
    if count >= i : # 그룹 인원이 공포도 이상이면
        result += 1 # 그룹 수
        count = 0 # 다음 그룹으로

print(result)
```
