<span style="color: #ffd33d">예제</span>
-----------

```python
import time

def loc_check(space, row, col) :
    if row < 1 or row > space :
        return True
    if col < 1 or col > space :
        return True

def LRUD(space, command, row, col) :
    if command == 'L' :
        drow = row
        dcol = col-1
    elif command == 'R' :
        drow = row
        dcol = col+1
    elif command == 'U' :
        drow = row-1
        dcol = col
    elif command == 'D' :
        drow = row+1
        dcol = col
    else :
        print("undifined command")
        
    if loc_check(space, drow, dcol) :
        print("ignore command : out of range")
        drow, dcol = row, col
    
    return drow, dcol

space = int(input("공간 크기 : "))
command_list = list(input("계획서 입력 : ").split())

s_t = time.time()

row, col = 1, 1

for c in command_list :
    row, col = LRUD(space, c, row, col)
    print(row, col)
    
print("*" * 50)
print(row, col)

e_t = time.time()
print("time :", e_t-s_t)
```
