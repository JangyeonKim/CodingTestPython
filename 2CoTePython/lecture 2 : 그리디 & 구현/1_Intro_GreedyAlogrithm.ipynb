- "현재 상황에서 지금 당장 좋은 것"을 고르는 방법



- 그리디 알고리즘은 최적의 해를 보장할 수 없을 때가 많다.
- 대부분 그리디 문제는 탐욕법으로 얻은 해가 최적의 해인 상황에서, 이를 추론할 수 있어야 풀리도록 출제.



<span style="color: #ffd33d">예제</span>
----------

##### <예제1 : 거스름돈>


문제 해결 아이디어 :   
Q : 가장 큰 단위부터 돈을 거슬러 주는 것이 최적의 해를 보장 하는 이유 ?   
A : "큰 단위가 항상 작은 단위의 배수"이므로 작은 단위들을 조합해 다른 해가 나올 수 없기 때문




```python
N = 1260 # 1260원

cnt = 0

for i in [500, 100, 50, 10] :
    if N == 0 :
        pass
    else :
        cnt += N // i # 동전 개수
        N %= i # 나머지
        
print(cnt)
```
