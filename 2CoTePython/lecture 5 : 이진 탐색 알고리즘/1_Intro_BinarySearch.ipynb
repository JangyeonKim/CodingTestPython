-----------



- 리스트가 정렬되어 있을 때 사용 가능



- 중간점이 두개일 경우(짝수개)에는 왼쪽 중간점을 선택
- 찾고자 하는 값이 중간점보다 작을 경우 중간점 오른쪽은 탐색할 필요 없음. 왼쪽만 탐색 진행




```python
# 이진 탐색 예시 코드

# 1. 재귀적 구현
def binary_search(array, target, start, end) :
    if start > end :
        return None
    mid = (start + end) // 2
    
    if array[mid] == target : # 중간점이 찾고자하는 값이면 중간점 인덱스 반환
        return mid
    elif array[mid] > target : # 중간점의 값이 찾고자하는 값보다 크면 왼쪽 확인
        return binary_search(array, target, start, mid-1)
    else : # 반대의 경우 오른쪽 확인
        return binary_search(array, target, mid+1, end)

# # 2. 반복문 구현
# def binary_search(array, target, start, end) :
#     while start <= end :
#         mid = (start + end) // 2
        
#         if array[mid] == target :
#             return mid
#         elif array[mid] > target :
#             end = mid - 1 # 끝 인덱스를 중간점 - 1로 변경
#         else :
#             start = mid + 1 # 시작 인덱스를 중간점 + 1로 변경
#     return None

n, target = list(map(int, input().split())) # 원소의 개수와 찾고자하는 값 입력받기
array = list(map(int, input().split())) # 전체 원소 입력받기

result = binary_search(array, target, 0, n-1) # 이진 탐색 수행 결과 출력
if result == None :
    print("원소가 존재하지 않습니다.")
else :
    print(result + 1) # 인덱스는 0부터 시작하므로 1 더하여 위치 출력
```

```python
from bisect import bisect_left, bisect_right

a = [1,2,4,4,8]
x = 4

print(bisect_left(a,x)) # 정렬된 리스트 a에 x를 삽입할 가장 왼쪽 인덱스 : 2 출력
print(bisect_right(a,x)) # 정렬된 리스트 a에 x를 삽입할 가장 오른쪽 인덱스 : 4 출력
```

```python
# bisect를 이용하여 "값이 특정 범위에 속하는 원소의 개수 구하기" 예시 코드

from bisect import bisect_left, bisect_right

# 값이 [left_value, right_value]인 데이터의 개수를 반환하는 함수
def count_by_range(a, left_value, right_value) :
    right_index = bisect_right(a, right_value)
    left_index = bisect_left(a, left_value)
    return right_index - left_index

a = [1,2,3,3,3,3,4,4,8,9]

print(count_by_range(a, 4, 4)) # 값이 4인 데이터 개수 출력 : 2
print(count_by_range(a, -1, 3)) # 값이 [-1, 3] 범위에 있는 데이터 개수 출력 : 6
```
