
- 최적화 문제 : 어떤 함수의 값을 가능한 한 크게 만들거나 작게 만드는 문제
- 이진 탐색을 이용하여 탐색 범위를 줄여가며 최적화 문제를 해결 가능

#### <문제 1 : 떡볶이 떡 만들기>




```python
# 내가 풀어본 풀이 : 이진 탐색을 적용하지 않았음

import time

N, M = map(int, input().split())

dduk_list = list(map(int, input().split()))

start_time = time.time()

diff_list = [0] * N

cut_len = max(dduk_list)
cut = 0 

while cut < M : # 잘린 길이의 합이 요청한 길이 M보다 같거나 커지면 반복 종료
    cut = 0
    cut_len -= 1
    
    for i in range(N) :
        if dduk_list[i] > cut_len :
            diff_list[i] = dduk_list[i] - cut_len
        else : 
            diff_list[i] = 0
        
    for i in range(N) :
        cut += diff_list[i]

end_time = time.time()

print(end_time - start_time)
print(diff_list)
print(cut_len)
```

```python
import time

n, m = list(map(int, input().split()))
array = list(map(int, input().split()))

start_time = time.time()

# 이진 탐색을 위한 시작점과 끝점 설정
start = 0
end = max(array)

result = 0

while start <= end :
    total = 0
    mid = (start + end) // 2 # 중간점 설정
    
    for x in array :
        if x > mid : # 잘랐을 때의 양 계산
            total += x-mid
    
    if total < m : # 양이 부족할 영우
        end = mid - 1 # 왼쪽 부분 탐색하여 더 많이 자를 수 있도록
    else :
        result = mid # result 기록
        start = mid + 1 # 오른쪽 탐색하여 최대한 덜 자를 수 있도록 탐색
        
end_time = time.time()
print(end_time - start_time)

print(result)
```

#### <문제 2 : 정렬된 배열에서 특정 수의 개수 구하기>




```python
from bisect import bisect_left, bisect_right

N, x = map(int, input().split())
array = list(map(int, input().split()))

def counter(x, array) : 
    left = bisect_left(array, x)
    right = bisect_right(array, x)
    
    result = right - left
    
    return result if result != 0 else -1 # 결과가 0이면 -1 출력, 아니면 결과 출력

print(counter(x,array))
```
