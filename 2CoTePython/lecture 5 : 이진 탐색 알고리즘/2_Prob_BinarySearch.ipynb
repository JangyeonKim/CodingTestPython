```python
#    :    

import time

N, M = map(int, input().split())

dduk_list = list(map(int, input().split()))

start_time = time.time()

diff_list = [0] * N

cut_len = max(dduk_list)
cut = 0 

while cut < M : #      M    
    cut = 0
    cut_len -= 1
    
    for i in range(N) :
        if dduk_list[i] > cut_len :
            diff_list[i] = dduk_list[i] - cut_len
        else : 
            diff_list[i] = 0
        
    for i in range(N) :
        cut += diff_list[i]

end_time = time.time()

print(end_time - start_time)
print(diff_list)
print(cut_len)
```

```python
import time

n, m = list(map(int, input().split()))
array = list(map(int, input().split()))

start_time = time.time()

#      
start = 0
end = max(array)

result = 0

while start <= end :
    total = 0
    mid = (start + end) // 2 #  
    
    for x in array :
        if x > mid : #    
            total += x-mid
    
    if total < m : #   
        end = mid - 1 #        
    else :
        result = mid # result 
        start = mid + 1 #        
        
end_time = time.time()
print(end_time - start_time)

print(result)
```

```python
from bisect import bisect_left, bisect_right

N, x = map(int, input().split())
array = list(map(int, input().split()))

def counter(x, array) : 
    left = bisect_left(array, x)
    right = bisect_right(array, x)
    
    result = right - left
    
    return result if result != 0 else -1 #  0 -1 ,   

print(counter(x,array))
```
