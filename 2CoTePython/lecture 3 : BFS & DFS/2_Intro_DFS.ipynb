<span style="color: #ffd33d">DFS 개요</span>
-----------

- 그래프의 깊은 부분부터 우선적으로 탐색
- 스택 자료구조를 이용



- DFS 동작 예시







```python
"""
DFS 구현 예제
"""

# DFS 함수 정의
def dfs(graph, v, visited):
    # 현재 노드를 방문 처리
    visited[v] = True
    print(v, end=' ')
    # 현재 노드와 연결된 다른 노드를 재귀적으로 방문
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)

# 각 노드가 연결된 정보를 리스트 자료형으로 표현(2차원 리스트)
graph = [
  [], # 노드의 인덱스가 1번부터 시작하는 경우가 많기 때문에 0번을 비워둠
  [2, 3, 8],
  [1, 7],
  [1, 4, 5],
  [3, 5],
  [3, 4],
  [7],
  [2, 6, 8],
  [1, 7]
]

# 각 노드가 방문된 정보를 리스트 자료형으로 표현(1차원 리스트)
visited = [False] * 9 # initialize : 모든 노드가 아직 방문되지 않은 상태
                      # 인덱스 0은 사용하지 않기 위해 하나 더 큰 크기로 생성

# 정의된 DFS 함수 호출
dfs(graph, 1, visited)
```
