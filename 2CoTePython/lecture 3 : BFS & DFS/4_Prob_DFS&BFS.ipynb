```python
def dfs(x, y) : 
    if x <= -1 or x >= N or y <= -1 or y >= M :
        return False #      
    
    #    
    if graph[x][y] == 0 :
        graph[x][y] = 1 #    
        # , , ,     
        dfs(x-1, y)
        dfs(x+1, y)
        dfs(x, y-1)
        dfs(x, y+1)
        return True
    return False

N, M = map(int, input().split())

graph = []
for _ in range(N) :
    graph.append(list(map(int, input()))) #   2  

count = 0 #       

#    
for i in range(N) : 
    for j in range(M) :
        if dfs(i,j) == True :
            count += 1
            
print(count)
```

```python
from collections import deque

# N, M     
n, m = map(int, input().split())
# 2     
graph = []
for i in range(n):
    graph.append(list(map(int, input())))

#      (, , , )
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

# BFS  
def bfs(x, y):
    # (Queue)   deque  
    queue = deque()
    queue.append((x, y))
    #    
    
    while queue:
        x, y = queue.popleft()
        #   4   
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            #      
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            #   
            if graph[nx][ny] == 0:
                continue
            #        
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1
                queue.append((nx, ny))
            print(queue)
    #      
    return graph[n - 1][m - 1]

# BFS   
print(bfs(0, 0))
```
