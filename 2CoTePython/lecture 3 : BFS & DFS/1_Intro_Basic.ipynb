#### <stack & queue>
- stack : LIFO(Last In First Out) 구조




- queue : FIFO(First In First Out) 구조
    - 리스트를 이용하여 구현할 수도 있지만, 시간복잡도가 높아 비효율적으로 동작. **deque 라이브러리** 이용





```python
'''
stack 구현 예제
'''

stack = []

# 5 - 2 - 3 - 7 - 삭제 - 1 - 4 - 삭제 

stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
stack.pop()
stack.append(1)
stack.append(4)
stack.pop()

print(stack[::-1]) # 최상단 원소부터 출력. 1, 3, 2, 5
```

```python
'''
queue 구현 예제
: 관행적으로 append()와 popleft() 사용. 데이터가 오른쪽으로 들어와 왼쪽으로 나감.
'''

from collections import deque

queue = deque() 

# 5 - 2 - 3 - 7 - 삭제 - 1 - 4 - 삭제 

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue) # 3, 7, 1, 4
```

#### <재귀 함수>


- DFS를 실질적으로 구현하고자할 때, 자주 사용되는 방법
- 문제 풀이에서 사용할 때는 재귀 함수의 종료 조건을 반드시 명시해야 함
    - 재귀 함수는 스택의 대표적인 예시



```python
'''
팩토리얼 구현 예제
'''

# 반복문으로 구현한 팩토리얼
def iterative_fac(n) :
    result =  1
    
    for i in range (1, n+1) :
        result *= i 
    
    return result

# 재귀적으로 구현한 팩토리얼
def recursive_fac(n) :
    if n <= 1 :
        return 1
    return n*recursive_fac(n-1)

print("iter :",iterative_fac(5),"recur :", recursive_fac(5))
```

```python
'''
유클리드 호제법 구현 예제
'''

def GCD(a, b) :
    if a % b == 0 :
        return b
    else :
        return GCD(b, a%b)
    
print(GCD(192, 162))
```
