#### <문제 : 두 배열의 원소 교체>





```python
import time

N, K = map(int, input().split())

A = list(map(int, input().split()))
B = list(map(int, input().split()))

start_time = time.time()

A.sort()
B.sort()

for i in range(K) :
    if A[i] < B[-1-i] :
        A[i], B[-1-i] = B[-1-i], A[i]
    else :
        break # A의 원소가 더 크거나 같아지면 반복문 탈출
    
end_time = time.time()

print(A)
print(B)
print(sum(A))

print(f"time : {end_time-start_time}")
```
